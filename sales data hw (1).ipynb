# Data Preprocessing Insights

1. The dataset initially contained date columns (order_date, ship_date) stored as object (string) type. 
These were converted into datetime format to enable time-based analysis.

2. The sales column was incorrectly stored as object due to formatting issues. 
It was cleaned and converted into a numeric type for statistical calculations and visualization.

3. A logical inconsistency was detected where a transaction had sales = 0 but quantity > 0, which violates real-world business logic. 
This record was removed to maintain data integrity.

4. No missing values or duplicate rows were found in the dataset.

5. Data types were verified to ensure numerical columns are suitable for analysis.

# Feature Engineering Insights

1. A new feature ship_delay was created to measure the number of days between order and shipment, helping analyze logistics efficiency.

2. profit_margin was generated to understand profitability relative to sales rather than absolute profit values.

3. high_value_order flag was introduced to identify orders above the median sales value, highlighting revenue-driving transactions.

4. discount_flag was created to differentiate between discounted and non-discounted orders.

5. Time-based features such as order_month and order_day were derived to analyze seasonal and weekly sales trends.

# EDA (Exploratory Data Analysis) Insights

1. Sales Distribution

Sales data is highly right-skewed.

Most orders have low-to-medium sales values, while a few very large transactions act as outliers and contribute heavily to total revenue.

2. Profit Distribution

Profit shows high variability with both positive and negative values.

Some transactions result in significant losses, while a few generate extremely high profits.

3. Shipping Cost

Shipping costs are generally low, but extreme high-cost shipments exist, which can reduce overall profitability.

4. Category-wise Sales

5.Technology category shows the highest variability and high-value transactions.

6. Office Supplies mostly contains low-value orders.

7. Furniture has moderate and more consistent sales values.

#Overall Pattern

Most numerical variables show right-skewed distributions with significant outliers.

A small proportion of orders drives a large share of revenue and profit.
